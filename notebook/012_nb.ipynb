```python
from scipy.spatial import distance

for i in X_train['fold'].unique():
    trn_group_city = X_train[X_train['fold'] != i].groupby('City')[['lat', 'lon']].median().reset_index()
    trn_dist = distance.cdist(trn_group_city[['lat', 'lon']], trn_group_city[['lat', 'lon']], metric='euclidean')
    trn_dist = pd.concat([trn_group_city[['City']], pd.DataFrame(trn_dist, columns=trn_group_city['City'])], axis=1)

    df_trn_near_city = {}
    for city in trn_dist.City:
        df_trn_near_city[city] = trn_dist[['City', city]].sort_values(city).City.to_list()
    df_trn_near_city = pd.DataFrame(df_trn_near_city).drop(0).reset_index(drop=True)
    print(df_trn_near_city.head())
```
